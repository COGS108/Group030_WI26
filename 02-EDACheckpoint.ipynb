## Research Question

How are official YouTube music videos, song tempo, and release seasons of Pop songs associated with achieving higher chart rankings in the Billboard Hot 100 Year-End Charts between 2015 and 2025? In this study, we will take a closer look at these top ranking songs’ calendar release dates, average BPMs, and whether they have official music videos to determine if these factors are correlated with their Billboard chart positions.



## Background and Prior Work

What determines the ranking of Pop songs in renowned music lists like the Billboard Hot 100 Year-End Charts? Through our investigation, we hope to identify potential factors that may be associated with chart positions of these top songs. Can a song’s BPM indicate a higher ranking? What about the presence of an official music video? Is it more likely for a song to gain more popularity if it is released in the summer over the winter? By answering these questions, we aim to explore what characteristics drive chart placements of Pop songs, one of the largest and most diverse genres in the music industry. 

In past research conducted by Nunes and Ordanini (2014)<a name="cite_ref-1"></a><sup><a href="#cite_note-1">1</a></sup>, they examined how the type of instrumentation in a Pop song affects their success on music charts. Specifically, they analyzed data from the Billboard Hot 100 Weekly Charts spanning 55 years, comparing number-one singles with songs that never climbed above rank 90 to determine what factors made number-one hits. Using Qualitative Comparative Analysis and Quantitative Regression techniques, Nunes and Ordanini found that the presence of background vocals was a necessary component consistently seen in number-one ranking Pop songs. Furthermore, their results suggested that in the 1960s and late 2000s, songs with relatively few instruments ranked higher in the charts. On the other hand, songs with relatively many instruments ranked higher during the 1970s–1990s. They also proposed that the songs that ranked number one on music charts like the Billboard Hot 100 were typically unique, set apart from the current era of music trends. These findings indicate that specific musical features can play a critical role in chart performance, which informs our investigation into how additional factors like tempo, release season, and official music video presence may also be associated with higher Billboard rankings.

In a similar study, Kim and Oh (2021)<a name="cite_ref-2"></a><sup><a href="#cite_note-2">2</a></sup> examined the relationship between certain audio features and a song’s success on the Billboard Hot 100 chart. They analyzed over 6,000 songs that appeared on the weekly charts between 1998 and 2016, and used regression analysis and predictive modeling to find patterns in the music that made a difference in chart performance. Their results showed that songs that felt more energetic and easy to dance to were more likely to reach the top 10. In particular, tracks with a steady and powerful beat often ranked higher on the charts. They also discovered that songs with a positive mood tended to perform better than those that sounded dark or melancholic. These uplifting songs seemed to connect more quickly with listeners and remain popular for longer periods of time. The study further indicated that timbre could also make an impact, as songs with clearer or more distinctive sounds were more likely to stand out and achieve higher chart rankings.

Both studies above help shape how we think about our own research. Nunes and Ordanini focused more on musical structure, like how background vocals or instruments relate to chart success over time. Kim and Oh looked closely at how things like danceability and happiness affect chart performance. While these studies help show that the features inside a song matter, they don’t explore outside factors such as what time of year the song came out, or whether it has an official music video that may have increased its popularity. That’s where our project comes in. We’re combining factors including tempo, release season, and the presence of music videos to see how these things work together to influence a song’s ranking on the Billboard Hot 100 Year-End charts from 2015 to 2025. And in contrast to these previous studies, which focused on weekly charts and short-term song success, our study examines long-term chart success for Pop songs by analyzing Year-End charts instead.

### Billboard Hot 100 Year-End Singles (2015-2025)

This dataset contains over a decade-long collection of top songs from various genres (R&B, HipHop, Pop, & Country) that were featured on the Billboard Hot 100 Year-End Charts between 2015 and 2025. As mentioned earlier, we obtained our data from Wikipedia, SongBPM, and Youtube, since most of our metrics, such as genre, BPM, music video presence, and release date, could not be curated only by looking at the Billboard Year-End Charts. 

#### Most Relevant Metrics: 
- **Chart Rank:** A pre-established  value from 1 to 100. In this context, a lower numerical value indicates a higher rank (e.g. A song that charted number 1 is considered higher ranking than a song that charted number 10).
- **BPM:** A beats per minute measurement that indicates the tempo of a song. To determine the tempo of Pop songs in particular, we will use the following ranges: a slow/calm, low-energy song is typically less than 100 BPM, a mid range tempo is 100–120 BPM, and a fast/upbeat, high-energy song has a tempo greater than 120 BPM.
- **Genre:** A column that generalizes and categorizes each song from its numerous sub-genres into a primary umbrella music genre, such as R&B, Hip-Hop, Pop, or Country.
- **Release Date:** Following the YYYY-MM-DD format, this column will help us classify the release season of each song in the dataset. Specifically, we will observe the "MM" portion of the date to determine whether the song was released in the winter (12, 01, 02), spring (03, 04, 05), summer (06, 07, 08), or fall (09, 10, 11).  
- **Music Video:** A binary representation (1/0) of the existence of an official YouTube music video, where 1 means Yes and 0 means No.

#### Major Concerns with the Dataset:

**1. Genre Simplification and Classification**  
Although contemporary music spans a wide range of genres and subgenres, our analysis reduces this diversity into four broad categories: Pop, Hip-Hop, R&B, and Country. Each song was assigned to only one dominant category, even if it blended or was influenced by multiple stylistic elements. Subgenres were grouped under their broader parent genres (e.g., Trap under Hip-Hop), while genres less directly aligned with these four categories (e.g., Rock or Disco) were classified according to their closest stylistic or commercial association.  
In cases involving fusion genres or derivative forms (e.g., Trap-Pop, Dance-Pop), or when a song spanned multiple broad categories (e.g., both Pop and Hip-Hop), classification was guided by the song’s dominant commercial positioning within the Billboard context rather than strict music-theoretical distinctions. In particular, because Pop as a category often functions as a broad commercial umbrella for music that appeals to a general audience, songs with cross-genre characteristics were categorized based on their primary market identity within the Billboard chart ecosystem.  
While this coarse-grained classification improves analytical consistency and tractability, it also inevitably introduces subjectivity and collapses nuanced stylistic distinctions into broader categories.  

**2. Version Ambiguity in Release Date Identification**  
Many songs exist in multiple versions, including remixes, collaborations, re-releases, edited editions, cover versions, etc. We would always choose the charted version of the song as the focal observation: When the original version charted, we collected data from the original version; and when a specific remix or collaboration version charted, we collected data from that specific charting version.  
To ensure operational consistency and reproducibility, if multiple identical-content releases occurred (re-release of the exact same song), we selected the final release date prior to the song’s major rise in popularity to maintain temporal consistency with its chart performance. Despite these efforts, release dates may not perfectly correspond to the period during which a song accumulated the performance contributing to its Year-End ranking.  

**3. Strict Operational Definition of Official Music Videos**  
Again, to ensure operational consistency and reproducibility, only videos explicitly labeled as “Official Music Video” or “Official Video” on YouTube were coded as present. Lyric videos, studio performances, live recordings, leaks, cancelled releases, and unofficial uploads were excluded.  
While this strict operational definition reduces classification ambiguity, it may underestimate informal, unofficial digital virality, platform-driven exposure, or user-generated content that contributed to a song’s popularity (e.g., a viral TikTok video with the clip of a song causes the song to also go viral).  

**4. Reverse Causality Between Music Videos and Chart Rankings**  
A central limitation concerns potential reverse causality between music video production and chart performance. 
In some cases, music videos are produced and released after a song has already begun gaining traction, functioning as a strategic supplement or reinforcement of rising popularity. Under such circumstances, it becomes difficult to determine whether music videos drive chart success, or whether chart success motivates video production.  
Because this is an observational study using aggregated Year-End chart data, our findings should therefore be interpreted as associative rather than causal.  

**5. Tempo Measurement and Cross-Source Variability**  
BPM values were obtained primarily from SongBPM and cross-checked from Tunebat when necessary. However, minor discrepancies across databases or differences between song versions (e.g., radio edits, remixes, half-time or double-time interpretations, and there are even songs with almost exact same title, cover, released on close dates, but has different BPM) may introduce measurement error.  
Although efforts were made to match tempo values to the specific charting version (as stated in 2. Version Ambiguity in Release Date Identification above, the data is always selected from the specific charting version), BPM should be understood as an approximate musical indicator rather than a perfectly fixed structural constant.  

**6. Release Season and Popularity Timing Misalignment**  
In the streaming era, songs frequently reach their peak popularity after months—or even years—of their initial release. As a result, release season may not fully capture the timing of chart-relevant success.  
Our seasonal classification reflects the release timing, rather than the true period during which cumulative performance contributes to the Billboard Year-End ranking. This temporal misalignment limits the strength of our inferences regarding seasonal effects.  

**7. Seasonal Evergreen Songs and Recurring Chart Effects**  
Certain cultural/ holiday-themed “evergreen” songs reappear on Year-End charts due to cyclical cultural consumption patterns. For example, All I Want for Christmas Is You and Rockin Around the Christmas Tree both have appeared repeatedly since 2020 (5 times in our dataset).  
These recurring songs may inflate winter representation and reflect annual listening traditions rather than the classical release dynamics of contemporaneous songs.  
While such songs were retained as part of the Billboard system to reduce subjectivity of data selection, they may introduce structural seasonality effects independent of release timing.

#### Data Cleaning Process


In the code below, we cleaned the dataset by removing any hidden spaces, filtering out non-Pop songs, replacing the Release Date column with the more relevant Release Season column, and finally resetting the index to ensure an organized structure.

```python
# Remove hidden spaces from column names
df.columns = df.columns.str.strip()

# Remove hidden spaces from the Genre text itself
df['Genre'] = df['Genre'].str.strip()

# Filter out non-Pop genres, only keeping singles that fall under the Pop genre in the dataset
df_pop = df[df['Genre'].str.contains('Pop', case=False, na=False)].copy()

# Replace Release Date column with Release Season column, which shows if a Pop song was released in the winter, spring, summer, or fall
seasons = []
for date_string in df_pop['Release Date']:
    month = date_string[5:7]    
    if month == '12' or month == '01' or month == '02':
        seasons.append('Winter')
    elif month == '03' or month == '04' or month == '05':
        seasons.append('Spring')
    elif month == '06' or month == '07' or month == '08':
        seasons.append('Summer')
    else:
        seasons.append('Fall')

# Add the new column and drop the old one
df_pop['Release Season'] = seasons
df_pop = df_pop.drop(columns=['Release Date'])

#Reset index
df_pop = df_pop.reset_index(drop=True)

# Save the clean data
df_pop.to_csv('data/02-processed/billboard_data_pop_only.csv', index=False)
```

**PLOT 4: Distribution of Top 25 Hits Across Release Seasons**

This 100% stacked bar plot looks at how the share of the Top 25 hits change across seasons. By splitting songs into two groups, Top 25 and 26-100, we can see if certain times of the year tend to produce more high performing tracks. The goal here is to find out if releasing timing plays a role in chart success or if Pop hits perform about the same no matter when they come out. 

```python
df_pop.loc[:, 'Rank Tier'] = pd.cut(df_pop['Rank'], 
                                    bins=[0, 25, 100], 
                                    labels=['Top 25', '26-100'])


contingency_table = pd.crosstab(df_pop['Release Season'], df_pop['Rank Tier'])

percent_table = contingency_table.div(contingency_table.sum(1), axis=0) * 100

season_order = ['Spring', 'Summer', 'Fall', 'Winter']
percent_table = percent_table.reindex([s for s in season_order if s in percent_table.index])

ax = percent_table.plot(kind='bar', 
                        stacked=True, 
                        figsize=(10, 6), 
                        color=['#0096FF', '#D3D3D3'],
                        edgecolor='white')

plt.title('Percentage of Top 25 Hits by Release Season', fontsize=15, pad=20)
plt.ylabel('Percentage (%)')
plt.xlabel('Season')
plt.xticks(rotation=0)
plt.legend(title='Rank Tier', bbox_to_anchor=(1.05, 1), loc='upper left')

for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy() 
    if height > 0:
        ax.text(x + width/2, 
                y + height/2, 
                f'{height:.1f}%', 
                ha='center', va='center', 
                color='white' if y < 50 else 'black', 
                fontweight='bold')

plt.tight_layout()
plt.show()
```

Looking at the results above, Fall holds the highest portion of Top 25 Pop hits at 34.9%, followed closely by Winter with 33.3%. This suggests that songs released during the colder months may perfrom slightly better on average in terms of chart placement. Spring, with 30.8%, maintains a moderate share of Top 25 tracks, while Summer shows the lowest proportion at 25.2%. Indicating that summer releases, despite being commonly associated with "hit" potential, may not always translate into top-tier chart success. Overall, these results imply that timing release might influence chart performance, but not in the way one might expect. Instead of Summer dominating the Pop charts, the data hints that Fall and Winter could be more favorable periods for higher-ranking songs. However, the differences across seasons are relatively small, suggesting that seasonality may play a minor role compared to other factors that drive a song's success. 

**PLOT 5: BPM Differences Between Top and Bottom 25 Pop Songs**

This box plot compares the BPM distributions for Pop songs in the Top 25 versus the Bottom 25 chart positions. By focusing on the extremes, we can see whether higher-ranking songs tend to have faster tempos on average, helping us explore whether tempo might be linked to chart success within the Pop genre. 

```python
df_extremes = df_pop.copy()

def get_extreme_tier(rank):
    if rank <= 25:
        return 'Top 25'
    elif rank >= 75:
        return 'Bottom 25'
    return None

df_extremes['Extreme Tier'] = df_extremes['Rank'].apply(get_extreme_tier)
df_plot = df_extremes.dropna(subset=['Extreme Tier'])

plt.figure(figsize=(8, 6))
sns.set_style("whitegrid")


ax = sns.boxplot(data=df_plot, 
                 x='Extreme Tier', 
                 y='BPM', 
                 hue='Extreme Tier',
                 order=['Top 25', 'Bottom 25'],
                 palette=['#6A0DAD', '#D3D3D3'], 
                 showmeans=True, 
                 meanprops={"marker":"d", "markerfacecolor":"white", "markeredgecolor":"black"},
                 legend=False)     

plt.title('BPM Distribution: Top 25 vs. Bottom 25', fontsize=15, pad=20)
plt.ylabel('Beats Per Minute (BPM)', fontsize=12)
plt.xlabel('Rank Tier', fontsize=12)

plt.tight_layout()
plt.show()
```

Looking at the box plot, Top 25 songs tend to have a slightly higher tempo, with BPMs ranging from around 100 to 137. In contrast, the Bottom 25 songs have a slightly lower BPM distribution starting around 97 and going up to 130. This suggests that faster-paced songs may have a modest advantage in reaching the top of the charts but the overlap in BPM ranges shows that tempo alone isn't a deciding factor. Some slower songs still break into the Top 25 and some faster songs remain in the Bottom 25. Overall, while higher BPM might give songs a slight edge, chart success is clearly influenced by multiple factors beyond just tempo.

#### 3. Observing Multivariate Interactions

**PLOT 7: Correlation Between Tempo (BPM) and Chart Success Across Release Seasons**


The following visualization is a faceted scatter plot comparing how 'Release Season' can influence the interactions between the 'BPM' and Billboard Chart 'Rank' of Pop songs. By implementing this multivariate approach in our analysis, we will test our hypothesis that higher BPMs have a stronger influence on a song's ability to reach top-tier chart positions (1-25) when the song is released in the summer compared to other seasons.

```python
interactions = sns.lmplot(data = df_pop, x = 'BPM', y = 'Rank', col = 'Release Season', hue = 'Release Season')

# Inverting y-axis so Rank 1 is at top and Rank 100 is at bottom
for ax in interactions.axes.flat:
    ax.set_ylim(105, 0)
    ax.set_yticks(np.arange(0, 110, 10))

    #highlights the Top 25 song ranks
    ax.axhline(25, ls='--', color='purple', alpha=0.7)
    
plt.show()
```

Looking at the results above, the summer chart exhibits the most noticeable positive slope, indicating that higher BPM may subtly increase the chances of a Pop song reaching higher ranks. In contrast, the winter releases chart shows a slightly downward slope and contains fewer song distributions past the 120 BPM mark, implying that a higher tempo might actually decrease a song's chances of ranking in the top-tier positions during colder months. Finally, fall and spring are both nearly horizontal slopes, suggesting that BPM may have little to no influence on song rank during these seasons. However, it is important to note the significant variability in the scatter plots along with the fact that none of the regression lines bypass the Top 25 threshold. This finding illustrates that although summer releases have the strongest positive pull regarding tempo influence on rank, BPM is more of a supporting factor than a sole determinant of success for Pop songs. These results align with the idea that Pop song success is a complex concept, with many collective factors apart from tempo that can influence a song's performance in elite charts like the Billboard Hot 100.

PLOT 9: Offical Music Video 
This is a line plot looking into the percentage of top pop songs with an Offical Music Video per year. 

```python
df_pop['video_binary'] = df_pop['Music Video'].map({
    True: 1, False: 0, 
    'Yes': 1, 'No': 0, 
    1: 1, 0: 0
})
video_trend = df_pop.groupby('Year')['video_binary'].mean() * 100
plt.figure(figsize=(12, 6))
sns.set_style("whitegrid")

sns.lineplot(
    x=video_trend.index, 
    y=video_trend.values, 
    marker='o', 
    color='#2ecc71', 
    linewidth=2.5
)

plt.title('Percentage of Pop Hits with an Official Music Video (2015-2025)', fontsize=14)
plt.ylabel('Percentage (%)', fontsize=12)
plt.ylim(0, 105)
plt.show()
```

**PLOT 10: Evolution of Median Pop Song Rank By Release Season from 2015-2025**

This lineplot tracks the charting performance of Pop songs over the last decade, categorized by their release season. By plotting the median rank for each year, we can identify long-term shifts in seasonal dominance and determine the validity of our hypothesis: whether summer releases truly are a consistent industry trend related with better chart success or simply a year-dependent, fluctuating factor.

```python
plt.figure(figsize = (14, 7))
sns.lineplot(data = df_pop, x = 'Year', y = 'Rank', hue = 'Release Season', hue_order = ['Winter', 'Spring', 'Summer', 'Fall'], marker = 'o', estimator='median', errorbar = None)

# Inverting y-axis so Rank 1 is at top and Rank 100 is at bottom
plt.gca().invert_yaxis() 

# Forcing plot to show all 11 years
plt.xticks(df_pop['Year'].unique(), rotation = 45)

plt.title('Seasonal Trajectory of Median Pop Song Rank (2015-2025)', fontsize = 16)
plt.ylabel('Median Rank', fontsize = 12)
plt.xlabel('Year', fontsize = 12)
plt.grid(True, linestyle = '--', alpha = 0.5)
plt.legend(title = 'Season', bbox_to_anchor = (1.05, 1), loc = 'upper left')

plt.show()
```

Before jumping into seasonal specifics, a relevant finding from this visualization regarding the Pop genre as a whole is that both 2016 and 2020 have median ranks successfully passing the top 20 position, suggesting that the Pop genre achieved the most success in Billboard chartings during those two years out of the entire decade. Moving onto observing the individual seasonal trends, the lineplot above illustrates that spring releases suprisingly have the most consistent trend, ranging a steady 30-50 median charting position over the years. Pop songs released in the winter, on the other hand, seem to be facing a gradual decline over the years (with the exception of some peak times, including 2016 and 2020). Its overall pattern indicates a drop from the top-tier ranks to the lower-middle ranks (45-70), especially after 2020. Fall releases had the most chart success in the years between 2018-2021, consistently placing a median Billboard rank within the top 40, but drastically dropped to the lower 70s during 2022 and remained in the bottom ranks there-after. One particularly interesting observation is that the summer Pop releases showed the most success during the years that the winter and fall releases were showing a sharp decline (2021-2022). However, considering the overall trends between all the seasons, Pop songs released in the summer fall mostly in the middle-tier range of 40-65, with the exception of its short peak in performance a little after the emergence of COVID-19. This insightful result indicates that summer releases may not have as strong of an influence on Billboard chart success as we initially assumed it to, pointing to the idea that some factors may only have short-term impacts on musical performance over the years.
